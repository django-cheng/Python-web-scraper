#### 使用 requests 函式庫發送 GET 請求並解析回應

```python
import requests

# 1. 對目標網址發送 GET 請求，並將完整的伺服器回應(一個封裝了請求結果的物件)存入 'res' 變數中
res = requests.get('https://httpbin.org/get')

# 我們一起觀察這個物件有甚麼屬性和功能

# 2. 檢查回應狀態碼(status_code)，200 代表請求成功
# 狀態碼可以告知我們請求的處理狀態
print(f'Status Code: {res.status_code}')

# 3. 查看 requests 函式庫自動判斷的編碼格式(encoding)
#    httpbin 預設使用 utf-8，所以 requests 通常能正確判斷
print(f'Encoding:{res.encoding}')

# (備註) 如果自動判斷錯誤導致 res.text 出現亂碼，
# 可取消以下這行的註解，手動指定正確的編碼
# res.encoding = 'utf-8'

# 4. 輸出解碼後的文字內容 (text;此處 httpbin 回傳的是 JSON 格式的文字)
print("\n--- Response Text ---")
print(res.text)
```

在 GET 請求中附加查詢參數 (Query String)

```python
# 1. 定義一個字典來存放要附加到 URL 的查詢參數 (Query String)
#    requests 函式庫會自動將其轉換為 "?key1=value1&key2=value2" 的格式
my_params = {
    'key1' : 'value1',
    'key2' : 'value2'
}
# 2. 在 GET 請求中，透過 params 參數傳遞我們的字典
#    requests 會自動將 my_params 附加到 URL 後面
res = requests.get('https://httpbin.org/get', params=my_params)

# 3. 透過 res.url 屬性，可以觀察到 requests 實際發出的完整 URL
#    這是一個很好的除錯技巧，用來確認我們的參數是否正確附加
print(f"Final URL: {res.url}")

# 4. 輸出伺服器回應的內容
#    觀察回傳的 JSON 中 "args" 欄位，可以看到伺服器確實收到了我們的參數
print("\n--- Response Text ---")
print(res.text)
```

在 GET 請求中自訂標頭 (Custom Headers)

```python
# 1. 定義一個字典來存放自訂的請求標頭 (Request Headers)
#    'user-agent' 是最常用的一個，用來告訴伺服器我們是什麼樣的瀏覽器或客戶端。
# 'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36'
my_headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36'
}

# 2. 在 GET 請求中，透過 headers 參數傳遞我們的自訂標頭字典
res = requests.get('https://httpbin.org/get', headers=my_headers)

# 3. 輸出伺服器回應的內容
#    觀察回傳的 JSON 中 "headers" -> "User-Agent" 欄位，
#    可以看到它已經變成了我們剛剛自訂的內容，而不是 requests 函式庫的預設值。
print("--- Response Text ---")
print(res.text)
```

在 GET 請求中夾帶 Cookie

```python
# 1. 定義一個字典來存放要傳送的 Cookie
#    字典的鍵 (key) 是 Cookie 的名稱，值 (value) 是其內容。
my_cookies = {
    'first_cookie' : 'hello', 'second_cookie' : 'world'
}

# 2. 在 GET 請求中，透過 cookies 參數傳遞我們的自訂 Cookie 字典
#    requests 函式庫會自動將這個字典轉換成正確的 Cookie 標頭格式。
res = requests.get('https://httpbin.org/get', cookies=my_cookies)

# 3. 輸出伺服器回應的內容
#    觀察回傳的 JSON 中 "headers" -> "Cookie" 欄位，
#    可以看到它包含了 "first_cookie=hello; second_cookie=world"，證明 Cookie 已成功送出。
print("--- Response Text ---")
print(res.text)
```
