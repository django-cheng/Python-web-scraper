```python
from bs4 import BeautifulSoup as bs
import requests as req
from pprint import pprint
import json
```

```python
url = 'https://www.ettoday.net/news/hot-news.htm'
res = req.get(url)
#soup = bs(...) : 正在進行「實例化 (Instantiation)」
soup = bs(res.text, 'lxml')

print(f"status code:{res.status_code}")  # 200
print("="*50)
print(soup.title)
print(soup.title.string)
print(soup.title.text)
#with open('ettoday-hot-news.html', 'w', encoding='utf-8') as f:
    #f.write(res.text)
```
