### 🔍 單點搜尋利器：`.find()`
此方法專用於鎖定 DOM 樹中「第一個」符合條件的節點，支援標籤名與屬性篩選。
不同於全數列出的 `.find_all()`，它僅回傳單一 `Tag` 物件或 `None`。
適合用於定位頁面標題、特定 ID 區塊或導航列等唯一性元素。


````python
# 使用find()方法取得標籤為<title>的元素
# find() 方法會回傳第一個符合條件的元素
print(soup.find('title'))
````

````python
# 練習 .find() 方法的 string 參數，以 HTML 標籤內所包夾的「純文字內容」(<tag>文字</tag>)作為搜尋條件
# 目標元素 : <dd>物聯網，常與 AI 結合成為 AIoT。</dd>
# 記得如果單獨使用 string 參數，回傳的會是「字串物件」，而非 HTML 標籤 
str = soup.find(string='物聯網，常與 AI 結合成為 AIoT。')

print(str)
print('=' * 40)
str_2 = soup.find('dd', string='物聯網，常與 AI 結合成為 AIoT。')
print(str_2)
````

````python
# 使用find()方法取得符合條件的第一個元素
# 找出第一個標籤為<span>且class屬性值為'definition'的元素

# 關鍵字 (Key = 'value')
span_class = soup.find('span', class_ = 'definition')
print(span_class)


print(' ')

# 字典法 (attrs = {'key' : 'value'})
span_class2 = soup.find('span', attrs = {'class' : 'definition'})
print(span_class2)


# 取得該元素的文字內容
print(f"文字內容:{span_class.text}")
````

````python
# 使用字典法
'''
練習對象:
<div class="viral" data-id="003" data-level="advanced">
<h3 class="term">Deep Learning (DL)</h3>
<span class="explanation">深度學習，模仿人腦神經網絡的多層次學習。</span>
</div>
'''
# 找出第一個標籤為<div>且class屬性值為'viral'，data-id屬性值為'003'，data-level屬性值為'advanced'的元素
viral_div = soup.find('div', attrs = {'class' : 'viral', 'data-id' : '003', 'data-level' : 'advanced'})
print(viral_div)


# 取得該元素的型態與文字內容

# 取得該元素的<span>子標籤的文字內容 (donotation)
span_explanation = soup.find('span', attrs = {'class' : 'explanation'})
print(f"文字內容:{span_explanation.text}")
````

````python
# 關鍵字法 (1)
'''
練習對象:
<span class="meaning">人工智慧，電腦模擬人類智慧的技術總稱。</span>
'''
#選擇符合指定class屬性值的標籤元素
#class 是 Python 定義類別的保留字（關鍵字），不能當變數名用，所以 BeautifulSoup 規定加個底線 class_ 來區分
tech = soup.find(class_ = 'meaning')
print(tech)
````

````python
# 關鍵字法 (2)
'''
練習對象:
<section id="chapter-1" class="section-block">
            <h2>第一章：核心概念與基礎</h2>
            ................
'''
#選擇符合指定id屬性值的標籤元素
chp1 = soup.find('section', id = 'chapter-1')
print(chp1)
````

````python
'''
練習對象:
<h3 class="term">Machine Learning (ML)</h3>
'''
# 根據標籤內的「文字內容」來搜尋元素
# 找出標籤為<h3>且文字內容包含'Machine Learning (ML)'的元素

ML = soup.find('h3', string = 'Machine Learning (ML)')
print(ML)
````

````python
'''練習對象:
<tr class="item active highlight">
                        <td class="character">方法</td>
                        <td class="phrase">Reinforcement Learning</td>
                        <td>強化學習，透過獎勵與懲罰機制學習。</td>
                    </tr>
'''
# 尋找同時有多個class屬性值的元素
# 「找出標籤為 <tr>，且其 class 屬性包含 'item' 或 'active' 或 'highlight' 其中任意一個的元素」

# 字典法
multi_dict = soup.find('tr', class_= {'item', 'active', 'highlight'})
print(multi_dict)
print(' ')

# 關鍵字法
multi_dict2 = soup.find('tr', class_ = 'item active highlight')
print(multi_dict2)
print(' ')

# 關鍵字法但是屬性值不照頁面順序寫
multi_dict3 = soup.find('tr', class_ = 'highlight item active')
print(multi_dict3)
````

### 🌲 全域搜尋：`find_all()`

> **`find_all()`** 是 BeautifulSoup 中用於遍歷 DOM 樹並檢索「所有」符合條件節點的核心方法。<br>它解決了 `find()` 僅能返回單一實例的限制，回傳的是一個包含所有匹配結果的列表 (`ResultSet`)。

當你需要取得頁面上**所有的** `<a>` 標籤，或是進行比單純「取得第一個特定元素」更複雜的批量資料抓取時，`find_all()` 是不可或缺的工具。

#### 函數原型與參數解析

```python
find_all(name, string, **kwargs)


````python
# 使用find_all()方法取得所有符合條件的元素
# find_all() 方法會回傳所有符合條件的元素，結果以列表形式
# **kwargs : 屬性過濾器 (Attributes Filter)
# 練習: 找出所有class屬性值為'term'的元素，並用遍例的方式印出
term = soup.find_all(class_ = 'term')

# print(f"term的型別為{type(term)}")
for t in term:
    print(t)

print(' ')

# with tag h3
trem2 = soup.find_all('h3', class_ = 'term')
for t in trem2:
    print(t)
````

````python
#find_all(string)
#Unicode 範圍 [\u4e00-\u9fa5] 代表中文字元
# r"" 代表原始字串，不進行跳脫字元處理
#雖然在 [\u4e00-\u9fa5] 中沒有反斜線問題，但養成寫 Regex 就加 r 的習慣是 Best Practice。
import re

chinese_string = soup.find_all(string = re.compile(r'[\u2e00-\u9fa5]'))
for c in chinese_string:
    print(c)
````
