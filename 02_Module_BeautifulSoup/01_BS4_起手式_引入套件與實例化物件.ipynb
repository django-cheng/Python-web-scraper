| 解析器 (Parser) | 典型用法 (Usage) | 優點 (Advantages) | 缺點 (Disadvantages) |
| :--- | :--- | :--- | :--- |
| **Python’s html.parser** | `BeautifulSoup(markup, "html.parser")` | 1. **標準庫內建**：無需安裝<br>2. 速度中等 | 1. 速度不如 lxml<br>2. 容錯力普通 |
| **lxml’s HTML parser** | `BeautifulSoup(markup, "lxml")` | 1. **極速**：C 語言實作<br>2. 容錯率高 | 需額外安裝 (`pip install lxml`) |
| **lxml’s XML parser** | `BeautifulSoup(markup, "xml")` | 1. **極速**<br>2. 唯一支援 XML | 需額外安裝 C 語言庫 |
| **html5lib** | `BeautifulSoup(markup, "html5lib")` | 1. **極度寬容**：模擬瀏覽器行為<br>2. 自動修補殘缺標籤 | 1. **非常慢**<br>2. 需額外安裝 |


### 🛠️ 技術選型：解析器決策流程
依據資料格式（XML/HTML）、效能需求及環境限制，選擇最適解析器的邏輯路徑。
一般情況下，**`lxml`** 是兼具速度與功能的最佳平衡點。
