### 🎯 CSS 選擇器精準定位：`.select_one()`
支援標準 **CSS Selector** 語法，能優雅解決多重 class 篩選與深層巢狀結構定位。
<br>
語法 `soup.select_one('selector')` 回傳**第一個**符合的 `Tag` 物件（或 `None`），比 `find()` 更具彈性。
<br>
這是處理複雜頁面結構的首選方案，可直接套用瀏覽器複製的 Selector 字串。


```python
# select_one()練習1
# 使用CSS選擇器 : 標籤選擇器
# 標籤選擇器 : 跟據指定標籤來選擇該元素
'''
練習對象
<a class="external-link" href="https://www.openai.com/" target="_blank">OpenAI 官網</a>
'''

a = soup.select_one('a')
print(a)
```

```python
# select_one() 練習 4
# 使用CSS選擇器 : 屬性選擇器
# 根據特定屬性(ex:href)來選擇該元素，可提供限定條件，如指定屬性值
'''練習對象:
<a class="external-link" href="https://www.openai.com/" target="_blank">OpenAI 官網</a>
'''

# href = soup.select_one('[href]')
# print(href)

# 完全匹配 - 指定屬性值必須完全相同才能選中

href2 = soup.select_one('[href = "https://www.openai.com/"]')
print(href2)
```

```python
# select_one() 練習 5
# 複合選擇器 (Compound Selectors)
# 基本公式 : 標籤名 + 修飾符 (中間無空格，代表AND)

'''
<a class="external-link" href="https://www.openai.com/" target="_blank">OpenAI 官網</a>
'''
# Tag + Class
# OpenAI = soup.select_one('a.external-link')
# print(OpenAI)

# Tag + Attribute
OpenAI2 = soup.select_one('a[href = "https://www.openai.com/"]')
print(OpenAI2)
```

### 批次搜尋：`select()`
- **定義**：回傳**所有**符合 CSS 選擇器的元素 **列表 (List)**，找不到則回傳 `[]`。
- **操作**：列表不能直接取 `.text`，必須使用 **`for` 迴圈** 遍歷每個元素。

```python
# select() 練習
# 使用CSS選擇器定位HTML中的'li'，並且找class屬性包含'post-item'的所有元素
'''
練習對象:
<li class="post-item"><span class="tag">#單位</span> <span class="content">Token</span> <small>(模型處理文字的最小單位，不完全等於單字。)</small></li>
                <li class="thread-post"><span class="tag">#參數</span> <span class="content">Parameter</span> <small>(模型內部可調整的變數，決定模型大小。)</small></li>
                <li class="post-item"><span class="label">#參數</span> <span class="content">Weights</span> <small>(權重，神經網絡中連接強度的數值。)</small></li>
                <li class="social-post"><span class="tag">#參數</span> <span class="content">Bias</span> <small>(偏差，神經元激活的閾值偏移量。)</small></li>
'''

# 將元素內的文字取出，放入一個list中
obj = soup.select('li.post-item')
# print(type(obj))

for o in obj:
    print(o.text)
```
